```kotlin
%use dataframe
// import statements
import de.fraunhofer.aisec.cpg.*
import de.fraunhofer.aisec.cpg.graph.*
import de.fraunhofer.aisec.cpg.passes.*
import de.fraunhofer.aisec.cpg.graph.edges.flows.*
import java.io.File

// set some variables
var file = File("../../test/resources/client.cpp")
// build a config
var config = TranslationConfiguration.builder()
    .topLevel(file.parentFile)
    .sourceLocations(file)
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.cxx.CLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.cxx.CPPLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.java.JavaLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.golang.GoLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.llvm.LLVMIRLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.python.PythonLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.typescript.TypeScriptLanguage")
    .optionalLanguage("de.fraunhofer.aisec.cpg.frontends.ruby.RubyLanguage")
    .loadIncludes(true)
    .addIncludesToGraph(true)
    .debugParser(true)
    .defaultPasses()
    .useUnityBuild(true)
    .useParallelPasses(false)
    .build()

var result = TranslationManager.builder().config(config).build().analyze().get()
result.functions.toColumn("functions")
```

```kotlin
fun Node.fancy(): String {
    return "${this.name} (${this.javaClass.simpleName})"
}

fun Dataflows<*>.toDataFrame(): DataFrame<Dataflow> {
    return this.toDataFrame() {
        "from" {
            "identifer" from { it.start.fancy() }
            "location" from { it.start.location }
        }
        "to" {
            "identifer" from { it.end.fancy() }
            "location" from { it.end.location }
        }
        "granularity" {
            "type" from { it.granularity.javaClass.simpleName }
            "detail" from {
                var granularity = it.granularity
                if(granularity is PartialDataflowGranularity) {
                    granularity.partialTarget?.fancy()
                } else {
                    ""
                }}
        }
    }
}

var sa = result.refs("sa")[3]
sa.prevDFGEdges.toDataFrame()
```
